### Classifying Gorgan vendors 

```python
#import libraries
import numpy as np
import pandas as pd
```

```python
# import vendor's database
df = pd.read_excel('D:/Job offers/Snapp food/Food Vendors.xlsx', sheet_name='Google map')
df
```

```python
df_copy=df
```

### data preprocessing

The columns of 'name', 'type', 'location', 'phone number' are not usefull to classify the vendors. Therefore they will be removed immediately.

```python
# removing the useless columns
df_copy=df_copy.drop(columns=['Name', 'Type','Location', 'Phone number','Opening time','Closing time'], axis=1)
df_copy
```

```python
# Changing nun numerical data to a numerical one

# Service options
Service_options=[]
for i in df_copy['Service options']:
    if i=='Dine-in & Takeout & Delivery':
        Service_options.append(123)
    elif i=='Takeout & Delivery':
        Service_options.append(23)
    elif i=='Dine-in & Takeout':
        Service_options.append(13)
    elif i=='Dine-in':
        Service_options.append(1)
    elif i=='Takeout':
        Service_options.append(2)
    elif i=='Delivery':
        Service_options.append(3)
    else:
        Service_options.append(0)
        
#Meal
Meal=[]
for i in df_copy['Meal']:
    if i=='full':
        Meal.append(123)
    elif i=='Lunch & Dinner':
        Meal.append(23)
    elif i=='Dinner': 
        Meal.append(3)
    else:
        Meal.append(0)

df_copy['Service options']=Service_options
df_copy['Meal']=Meal
df_copy
```

To fill work time's gap, we can replace the average value of work time per each category of meal; 
i.e. if a meal was recorded as 23 it means lunch and Dinner was served by its vendor, 
therefore we replace average amount of work time over vendors with 23 meal tyopes for filling its work time

```python
# filling the null values of work time 

#finding the average values per each Meal's category
Meal_mean=df_copy.groupby(['Meal']).mean()
Meal_mean
```

```python
# finding null cells of work time column
df_nall=df_copy[df_copy['Work time'].isna()]
df_nall
```

Since all null cells are related to Meal with 23 value, we just need to replace these cells with value of 12.96 which is the average value of work time over Meals with 23 value.

```python
#finding indexes of null cells
index=df_nall['Work time'].index.values
index
```

```python
#replacing null cell with 
for i in index:
    df_copy.iloc[i,4]=12.96
df_copy
```

```python
df_copy
```

To classify the candidate vendors of Q1, the column of rating is selected as target label and it will be divide into level '1' per rating equal or more than 4.5, and level '2' in the otherwise.

```python
# defining quality level of vendors for target lable 'rating'
rating=[]
for i in df_copy['rating']:
    if i>=4.5:
        rating.append(1)
    else:
        rating.append(2)
df_copy['rating']=rating
df_copy
```

The procedure of preprocessing just finished

### Classifier model 

'Rating' column is selected as target lable, and other remained columns are used as featurs

### Naive Bayes classifier

Since we have a small database, it is better to utilize from Naive Bayes classifier to classify the candidate vendors

```python
#import neccessary libraries
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
#devide data columns into features and target 
features = ['Closness degree to the center','Service options','Meal','Work time','reviewers comments']
target = 'rating'
```

```python
#Train-test split 
X_train, X_test, y_train, y_test = train_test_split(df_copy[features], df_copy[target], test_size=0.30, random_state=42)
```

```python
#create Naive bayes model
model = GaussianNB()
```

```python
#learn the model by train x and y data
logistic_classifier=model.fit(X_train, y_train)
```

```python
# prediction
predictions = model.predict(X_test)
y_test=y_test.to_frame('real')
y_test['prediction']=predictions
y_test
```

```python
#predic y_train
y_train_pred=model.predict(X_train)
y_train_pred
```

```python
#Create a comparison dataframe of predictions and real data
y_train=y_train.to_frame('real')
y_train['prediction']=y_train_pred
y_train
```

```python
#Calculate the accuracy of the model
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
# prepare the cross-validation procedure
cv = KFold(n_splits=5, random_state=1, shuffle=True)
# create model
model = GaussianNB()
# evaluate model
scores = cross_val_score(model, df_copy[features], df_copy[target] , scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
